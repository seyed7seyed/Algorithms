# DFS: Depth First Search
 - **O( V + E )**
 - **finds all friends of a node**

```python
def dfs( graph, node, visited=set(), Listed=[] ): 
    
    if node not in visited:
        visited.add(node)
        Listed.append(node)
        
        for neighbour in graph[node]:
            dfs( graph, neighbour, visited, Listed )
            
    return Listed
```

```python
# graph = { node: list of nighbors }

graph = {
  '2' : [ ],
  '3' : ['2','4'],
  '4' : ['8'],
  '5' : ['3','7'],
  '7' : ['8'],
  '8' : [ ]
}
print("Following is the Depth-First Search")
dfs( graph, '7', set(), [] )
```

# BFS: Breadth First Search
 - **O( V + E )**
 - **finds all friends of a node**

```python
def bfs(graph, node, visited=[], queue=[], Listed=[]): 
    
    visited.append(node)
    queue.append(node) 

    while queue:        
        m = queue.pop(0) 
        Listed.append(m) 
        
        for neighbour in graph[m]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)
                
    return Listed
```

```python
# graph = { node: list of nighbors }

graph = {
  '2' : [ ],
  '3' : ['2','4'],
  '4' : ['8'],
  '5' : ['3','7'],
  '7' : ['8'],
  '8' : [ ]
}
print("Following is the Breadth-First Search")
bfs( graph, '3', [], [], [] )
```
