# Dynamic

```python
def dynamic_fibonacci( n ):
    if n<=1:
        return n
    else:
        a,b = 0,1
        for i in range(n-1):
            c = a + b
            a = b
            b = c
        return( c )
            
dynamic_fibonacci( 1200 )
```

```python
def dynamic_fibonacci_print( n ):
    a = 0
    b = 1
    if n<=1:
        print( a, b, end=' ' )
    else:
        print( a, b, end=' ' )
        for i in range(n-1):
            c = a + b
            a = b
            b = c
            print( c, end=' ' )
            
dynamic_fibonacci_print(10)
```

# Recursive

```python
def recursive_fibonacci( n ):
    if n<=1:
        return n
    else:
        return recursive_fibonacci(n-1) + recursive_fibonacci(n-2)
    
print( recursive_fibonacci(10) )
```

# Algebraic

```python
def Algebraic_fibonacci( n ):
    return round( (0.447)*(1.618)**(n) - 0.447*(-0.618)**(n) )

Algebraic_fibonacci( 1200 )
```
